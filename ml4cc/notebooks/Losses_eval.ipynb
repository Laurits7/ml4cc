```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep
hep.style.use(hep.styles.CMS)
```

```python
# d = pd.read_csv('/home/laurits/ml4cc/training-outputs/20250103_FCC_350_epoch/logs/peakFinding/version_0/metrics.csv')
d = pd.read_csv('/home/laurits/ml4cc/training-outputs/20250109_CEPC_150_epoch/logs/peakFinding/version_0/metrics.csv')
# d = pd.read_csv('/home/laurits/ml4cc/training-outputs/20250102_CEPC_peakFinding//logs/peakFinding/version_0/metrics.csv')
d2 = pd.read_csv('/home/laurits/ml4cc/training-outputs/20250103_FCC_350_epoch/logs/peakFinding/version_0/metrics.csv')
# /home/laurits/ml4cc/training-outputs/20250109_CEPC_150_epoch/logs
```

```python
val_loss = np.array(d['val_loss'])
val_loss2 = np.array(d2['val_loss'])
# train_loss = np.array(d['train_loss'])
```

```python
val_loss = val_loss[~np.isnan(val_loss)]
val_loss2 = val_loss2[~np.isnan(val_loss2)]
# train_loss = train_loss[~np.isnan(train_loss)]
```

```python
min_idx = np.argmin(val_loss)
min_loss = val_loss[min_idx]
min_epoch = np.arange(len(val_loss))[min_idx]

min_idx2 = np.argmin(val_loss2)
min_loss2 = val_loss2[min_idx2]
min_epoch2 = np.arange(len(val_loss2))[min_idx2]
```

```python
# plt.scatter(np.arange(len(val_loss)), val_loss, label="val_loss [Adam]", color='blue')
# plt.plot(train_loss, label="train_loss [Adam]", ls="--", color='r')
plt.plot(val_loss, label="val_loss [AdamW]", ls="--", color='b')
plt.plot(val_loss2, label="val_loss2 [AdamW]", ls="--", color='r')
plt.scatter(min_epoch2, min_loss2, color='r')
plt.scatter(min_epoch, min_loss, color='b')
plt.legend()
plt.yscale('log')
plt.grid()
plt.ylabel('MSE loss [a.u.]')
plt.xlabel('epoch')
plt.xlim(0, 350)
```
